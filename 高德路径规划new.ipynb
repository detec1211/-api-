```python
import requests
import folium
import requests
import pypinyin
import json
 
'''
获取地址坐标
'''
 
def pinyin(word):
    s = ''
    for i in pypinyin.pinyin(word, style=pypinyin.NORMAL):
        s += ''.join(i)
    return s
 
def get_city_location(name):
    url = "https://restapi.amap.com/v3/geocode/geo?address={}&output=json&key=6ea82363a4d41501919645708c190593".format(name)

    res_text = requests.get(url).text
 
    retval = json.loads(res_text).get('status')
 
    city_list = []
    if retval == '1':
        city_name = json.loads(res_text).get('geocodes')[0].get('city')
        longitude = float(json.loads(res_text).get('geocodes')[0].get('location').split(',')[0])
        latitude = float(json.loads(res_text).get('geocodes')[0].get('location').split(',')[1])
 
        tmpl = {
            "key": pinyin(city_name),
            "latitude": latitude,
            "longitude": longitude,
            "name": city_name
        }
 
        city_list.append(tmpl)
        #return json.dumps(city_list, ensure_ascii=False)
        return (longitude, latitude)
    return None


api_key = '6ea82363a4d41501919645708c190593'  # 替换为您在高德开放平台获得的API Key

def get_directions(start_latitude, start_longitude, end_latitude, end_longitude):
    url = 'https://restapi.amap.com/v3/direction/walking'  # 步行路径规划API的URL

    params = {
        'key': api_key,
        'origin': f'{start_longitude},{start_latitude}',  # 起始点经纬度
        'destination': f'{end_longitude},{end_latitude}',  # 终点经纬度
        'output': 'json'
    }

    response = requests.get(url, params=params)
    data = response.json()

    if data['status'] == '1' and int(data['count']) > 0:
        path = data['route']['paths'][0]['steps']
        coordinates = []
    
        for step in path:
            polyline = step['polyline']
            points = polyline.split(';')

            for point in points:
                lng, lat = point.split(',')
                coordinates.append((float(lat), float(lng)))
        print(coordinates)
        return coordinates

    return None


def adjust_coordinates(coordinates):
    adjusted_coordinates = []

    for coordinate in coordinates:
        lat, lng = coordinate
        adjusted_lat = lat + 0.0028
        adjusted_lng = lng - 0.005
        adjusted_coordinates.append((adjusted_lat, adjusted_lng))

    return adjusted_coordinates

def display_directions(coordinates):
    map = folium.Map(location=[coordinates[0][0], coordinates[0][1]], zoom_start=12)  # 创建地图

    # 添加起点和终点
    folium.Marker(
        location=[coordinates[0][0], coordinates[0][1]],  # 起点经纬度
        popup='起点',
        icon=folium.Icon(color='green', icon='play')
    ).add_to(map)

    folium.Marker(
        location=[coordinates[-1][0], coordinates[-1][1]],  # 终点经纬度
        popup='终点',
        icon=folium.Icon(color='red', icon='stop')
    ).add_to(map)

    # 添加路径线
    folium.PolyLine(
        locations=coordinates,
        color='blue',
        weight=2
    ).add_to(map)

    # 添加离散点
    for coordinate in coordinates:
        folium.CircleMarker(
            location=[coordinate[0], coordinate[1]],
            radius=5,
            color='orange',
            fill=True,
            fill_color='orange'
        ).add_to(map)

    map.save('map_walking.html')  # 保存地图为HTML文件
start_longitude,start_latitude = get_city_location("坪山人民医院")
end_longitude,end_latitude = get_city_location("深圳技术大学")


path_coordinates = get_directions(start_latitude, start_longitude, end_latitude, end_longitude)

if path_coordinates is not None:
    adjusted_coordinates = adjust_coordinates(path_coordinates)
    display_directions(adjusted_coordinates)
else:
    print('路径规划失败')
```
